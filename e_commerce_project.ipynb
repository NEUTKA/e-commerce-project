# Задача 4 
# Сколько у каждого из пользователей в среднем покупок в неделю (по месяцам)?

```python
# Для начала соеденим две таблицы olist_customers_datase.csv и olist_orders_dataset.csv
merged_df_of_step4 = pd.merge(olist_customers_dataset, olist_orders_dataset, on='customer_id') 
```

```python
merged_df_of_step4
```

```python
# Перобразуем столбец order_purchase_timestamp в формат datetime так как именно доставленные заказы можно считать 
# купленными.
merged_df_of_step4.order_approved_at = pd.to_datetime(merged_df_of_step4.order_purchase_timestamp)
```

```python
merged_df_of_step4.dtypes
```

```python
# Преобразуем столбец order_approved_at в месяцы и годы Для того чтобы потом сгруппировать по ним покупки
merged_df_of_step4['Month'] = merged_df_of_step4.order_approved_at.dt.month
merged_df_of_step4['Year'] = merged_df_of_step4.order_approved_at.dt.year
```

```python
# отфильтруем таблицу на покупки. Товар куплен если был доставлен Клиенту.
merged_df_of_step4 = merged_df_of_step4.query('order_status == "delivered"')
merged_df_of_step4
```

```python
# Есть пустые значения. Есть заказы которые не были переданы в службу доставки(2) и которые не были доставлены(8). 
# Избавимся от них
merged_df_of_step4.isna().sum()
```

```python
# Избавимся от пустых значений 
merged_df_of_step4 = merged_df_of_step4.dropna()
merged_df_of_step4
```

```python
# Подсчет количества заказов каждого пользователя в каждом месяце
monthly_purchases = merged_df_of_step4.groupby(['customer_unique_id', 'Year', 'Month'], as_index=False) \
    .agg({'customer_id': 'count'}) \
    .rename(columns={'customer_id': 'purchase_count'})
```

```python
monthly_purchases
```

```python
# Функция для подсчета количества недель в месяце
def weeks_in_month(year, month):
    days_in_month = monthrange(year, month)[1]
    return days_in_month / 7
```

```python
# Применение функции к каждому ряду для подсчета количества недель
monthly_purchases['weeks_in_month'] = monthly_purchases.apply(lambda row: weeks_in_month(row['Year'], row['Month']), axis=1)
```

```python
# Округление значений в столбце weeks_in_month до второго знака.
monthly_purchases['weeks_in_month'] = monthly_purchases.weeks_in_month.round(2)
```

```python
monthly_purchases
```

```python
# Рассчет среднего количества покупок в неделю для каждого пользователя в каждом месяце
monthly_purchases['average_purchases_per_week'] = monthly_purchases['purchase_count'] / monthly_purchases['weeks_in_month']
```

# Ответ 
# Теперь найдем, сколько у каждого из пользователей в среднем покупок в неделю (по месяцам)

```python
monthly_purchases
```

# Задача 5
## Выполните когортный анализ пользователей. В период с января по декабрь выявите когорту с самым высоким retention на 3-й месяц.

```python
# Объединяем два DataFrame по полю 'customer_id'
merged_df_step5 = olist_orders_dataset.merge(olist_customers_dataset, on='customer_id')
```

```python
# Убираем строки с пропущенными значениями в столбце 'order_approved_at'
for_kogort = merged_df_step5.dropna(axis='index', how='any', subset=['order_approved_at'])
```

```python
for_kogort.isna().sum()
```

```python
# Находим дату первой покупки для каждого уникального клиента и добавляем новый столбец 'first_purchase'
for_kogort['first_purchase'] = for_kogort.groupby('customer_unique_id')['order_approved_at'].transform('min')
```

```python
# Преобразуем столбцы с датами в формат datetime
for_kogort['order_purchase_timestamp'] = pd.to_datetime(for_kogort['order_purchase_timestamp'])
for_kogort['order_approved_at'] = pd.to_datetime(for_kogort['order_approved_at'])
for_kogort['first_purchase'] = pd.to_datetime(for_kogort['first_purchase'])
```

```python
# Преобразуем даты в формат 'YYYY-MM' для агрегирования по месяцам
for_kogort.order_purchase_timestamp = for_kogort.order_purchase_timestamp.dt.strftime('%Y-%m')
for_kogort.order_approved_at = for_kogort.order_approved_at.dt.strftime('%Y-%m')
for_kogort.first_purchase = for_kogort.first_purchase.dt.strftime('%Y-%m')
```

```python
for_kogort.head()
```

```python
# Фильтруем данные по статусу заказа и периоду первой покупки (2017 год)
for_kogort_table = for_kogort.query('order_status == ["processing", "shipped", "delivered"]')\
    .query("'2017-01' <= first_purchase <= '2017-12'")\
    .groupby(['first_purchase', 'order_approved_at'],as_index=False).agg({'customer_unique_id': 'nunique'})
```

```python
# Преобразуем столбцы с датами в формат периода 'месяц'
for_kogort_table.first_purchase = pd.to_datetime(for_kogort_table.first_purchase).dt.to_period("m")
for_kogort_table.order_approved_at = pd.to_datetime(for_kogort_table.order_approved_at).dt.to_period("m")
```

```python
# Вычисляем разницу в месяцах между датами 'order_approved_at' и 'first_purchase'
for_kogort_table['diff'] = (for_kogort_table.order_approved_at - for_kogort_table.first_purchase).apply(attrgetter("n"))
```

```python
# Строим сводную таблицу, где строки - месяц первой покупки, столбцы - разница в месяцах, значения - число уникальных клиентов
for_kogort_table_finish = for_kogort_table.pivot(index='first_purchase', columns = 'diff', values = 'customer_unique_id')
```

```python
# Рассчитываем коэффициенты удержания клиентов, деля все значения на значение первого месяца для каждой когорты
retention = for_kogort_table_finish.div(for_kogort_table_finish.iloc[:,0], axis = 0)
# Настраиваем отображение итоговой таблицы
final_table = (retention
            .style
            .set_caption('Retention by cohort')  # добавляем подпись
            .background_gradient(cmap='viridis')  # раскрашиваем ячейки по столбцам
            .highlight_null('white')  # делаем белый фон для значений NaN
            .format("{:.2%}", na_rep=""))  # числа форматируем как проценты, NaN заменяем на пустоту
# Отображаем итоговую таблицу удержания
final_table
```
